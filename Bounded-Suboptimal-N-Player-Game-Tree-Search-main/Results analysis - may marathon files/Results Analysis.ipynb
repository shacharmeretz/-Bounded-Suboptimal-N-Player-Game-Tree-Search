```python
def plot_scatter3(df, epsilons):
    df1 = df.copy(deep = True)
    df1['Algorithm'] = 'Bounded suboptimal'
    df1['Val'] = df1['ER_Bounded']
    df2 = df.copy(deep = True)
    df2['Algorithm'] = 'Shallow'
    df2['Val'] = df1['ER_Shallow']
    df = pd.concat([df1, df2])
    depths = df['max_depth'].unique()   
    fig, axs = plt.subplots(len(depths), figsize=(15,10), )
    #fig.suptitle('Expansion Rate vs. Epsilon', fontsize=20)
    for j in range(len(depths)) :
        depth = depths[j]
        ax = sns.lineplot(ax = axs[j], x = 'epsilon', y = 'Val', data = df[df['max_depth']==depth], hue = 'Algorithm', ci = None, palette=pairs[j*2:j*2+2], marker='o')
        ax.set_xlabel(' ',labelpad=20)
        ax.set_ylabel('Expansion Rate', rotation=90, fontsize=20, labelpad=20)
        ax.set_ylim(0, 1)
        ax.set_title(f'Depth = {depth}', fontsize=20)
        ax.set_xticks(list(set(df['epsilon'])))
        plt.setp(ax.get_legend().get_texts(), fontsize='15') 
        plt.setp(ax.get_legend().get_title(), fontsize='15')
        plt.setp(ax.get_xticklabels(), fontsize=14)         
        plt.setp(ax.get_yticklabels(), fontsize=14)    
    
        
    axs[-1].set_xlabel('Epsilon', rotation=0, fontsize=20, labelpad=10)
    fig.tight_layout() 
    plt.show()
print('Max sum = 64, Branching factor = 5, NPlayers = 3')
plot_scatter3(output,output['epsilon'].unique())
```

```python
def plot_scatter5(df, epsilons):
    depths = df['max_depth'].unique()   
    fig, axs = plt.subplots(len(depths), figsize=(15,10))
    fig.suptitle('Result Suboptimality vs. Epsilon', fontsize=20)
    x = df['epsilon']
    y_bound = df['Bound_Result_Gap']
    
    for j in range(len(depths)) :
        depth = depths[j]
        shallow_er_means = [4,8,16,24,32]
        bounded_er_means = []
        for i in df['epsilon'].unique():
            query = 'max_depth=={} '.format(depth)
            temp1 = df.query(query)
            query = 'epsilon=={} '.format(i)
            temp1 = temp1.query(query)
            bounded_er_means.append(np.mean(temp1['Bound_Result_Gap']))
        axs[j].plot(df['epsilon'].unique(), shallow_er_means, label='Max possible suboptimality', linestyle='-.')
        axs[j].plot(df['epsilon'].unique(), bounded_er_means, label='Actual result suboptimality. depth = {}'.format(depth), linestyle='--')
        axs[j].set_xlabel('epsilon', rotation=0, fontsize=15, labelpad=10)
        axs[j].set_ylabel('result suboptimality', rotation=90, fontsize=15, labelpad=20)
        axs[j].set_ylim(0, 35)
        axs[j].legend(loc='upper right', bbox_to_anchor=(1.42, 1.015),prop={'size': 14})
    
    plt.show()
print('Max sum = 64, Branching factor = 5, NPlayers = 3')
plot_scatter5(output,output['epsilon'].unique())
```

```python
def read_files(dir_path, files_list, epsilons):
    dfs = []
    opt_df = pd.read_csv(dir_path+files_list[0], nrows=files_rows)
    opt_df = opt_df[['final_hands_p4', 'final_hands_other', 'avg_depth']]
    opt_df.rename({'final_hands_p4': 'final_hands_p4_Opt', 'final_hands_other': 'final_hands_other_Opt', 'avg_depth': 'avg_depth_Opt'}, axis=1, inplace=True)
#     opt_df['won_Opt'] = opt_df.apply(lambda r:True if (r['final_hands_p4_Opt'] > max(r['final_hands_other_Opt'])) else False, axis =1)
    i=1
    for eps in epsilons:
        eps_df = pd.read_csv(dir_path+files_list[i], nrows=files_rows)
        eps_df = eps_df[['final_hands_p4', 'final_hands_other', 'avg_depth']]
#         eps_df['won'] = eps_df.apply(lambda r:True if (r['final_hands_p4'] > max(r['final_hands_other'])) else False)
        merged_df = eps_df.merge(opt_df, left_index=True, right_index=True)
        merged_df['p4_ratio'] = merged_df.apply(lambda r: (r['final_hands_p4']/r['final_hands_p4_Opt'])-1, axis=1)
        merged_df['depth_ratio'] = merged_df.apply(lambda r: (r['avg_depth']/r['avg_depth_Opt'])-1, axis=1)
        merged_df['epsilon'] = [eps] * eps_df.shape[0]
        dfs.append(merged_df)
        i+=1
    return dfs
```

```python
df = pd.read_csv(data_path)
df = df.rename(columns={'Unnamed: 0': 'game'})
df
```

```python
import ast

def tournament_analysis(raw_df):    
    res_df = pd.DataFrame(columns = ['Game', 'Player', 'Total_nodes', 'Depth', 'Score', 'Position'])
    for idx, row in raw_df.iterrows():
        players = ast.literal_eval(row['players_names'])
        total_nodes = ast.literal_eval(row['total_nodes'])
        depth = ast.literal_eval(row['avg_depth'])
        all_scores = ast.literal_eval(row['final_hands'])
        for i in range(len(players)):
            player_score = all_scores[i]
            position = len([score for score in all_scores if score > player_score]) + 1
            entry = {'Game': row['game'], 'Player': players[i],
                      'Total_nodes': total_nodes[i], 'Depth': depth[i],
                      'Score': player_score, 'Position': position}
            res_df = res_df.append(entry, ignore_index=True)
    return res_df
new_df = tournament_analysis(df)
new_df    
```
