```python
path = r'C:/Users/pbiton/OneDrive - Intel Corporation/Desktop/final_proj_exp/final_result_25/'
paranoid_dir = path + 'paranoid/'
shallow_dir = path + 'shallow/'
shallow_files = ['game30_epsilon_0.csv', 'game25_epsilon_2.csv', 'game30_epsilon_4.csv', 'game30_epsilon_8.csv','game25_epsilon_16.csv']
import pandas as pd
epsilons = [2, 4, 8, 16]
files_rows = 25
```

```python
def read_files(dir_path, files_list, epsilons):
    dfs = []
    opt_df = pd.read_csv(dir_path+files_list[0], nrows=files_rows)
    opt_df = opt_df[['final_hands_p1', 'final_hands_other', 'avg_depth']]
    opt_df.rename({'final_hands_p1': 'final_hands_p1_Opt', 'final_hands_other': 'final_hands_other_Opt', 'avg_depth': 'avg_depth_Opt'}, axis=1, inplace=True)
#     opt_df['won_Opt'] = opt_df.apply(lambda r:True if (r['final_hands_p1_Opt'] > max(r['final_hands_other_Opt'])) else False, axis =1)
    i=1
    for eps in epsilons:
        eps_df = pd.read_csv(dir_path+files_list[i], nrows=files_rows)
        eps_df = eps_df[['final_hands_p1', 'final_hands_other', 'avg_depth']]
#         eps_df['won'] = eps_df.apply(lambda r:True if (r['final_hands_p1'] > max(r['final_hands_other'])) else False)
        merged_df = eps_df.merge(opt_df, left_index=True, right_index=True)
        merged_df['p1_ratio'] = merged_df.apply(lambda r: (r['final_hands_p1']/r['final_hands_p1_Opt'])-1, axis=1)
        merged_df['depth_ratio'] = merged_df.apply(lambda r: (r['avg_depth']/r['avg_depth_Opt'])-1, axis=1)
        merged_df['epsilon'] = [eps] * eps_df.shape[0]
        dfs.append(merged_df)
        i+=1
    return dfs
```

```python
def compare_dfs_final_score(df1, df2):
    diff = {}
    for i in range(df1.shape[0]):
        if not (df1['final_hands_p1'].iloc[i] == df2['final_hands_p1'].iloc[i] and df1['final_hands_other'].iloc[i] == df2['final_hands_other'].iloc[i]):
            diff[i] = df2['final_hands_p1'].iloc[i]-df1['final_hands_p1'].iloc[i]
            print(i)
    return diff
```

```python
shallow_df = read_files(shallow_dir, shallow_files, epsilons)
concat_df = pd.concat([shallow_df[0],shallow_df[1], shallow_df[2], shallow_df[3]])
```
