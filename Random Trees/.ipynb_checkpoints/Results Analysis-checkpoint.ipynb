```python
def plot_scatter3(df, epsilons):
    df1 = df.copy(deep = True)
    df1['Algorithm'] = 'Bounded suboptimal'
    df1['Val'] = df1['ER_Bounded']
    df2 = df.copy(deep = True)
    df2['Algorithm'] = 'Shallow'
    df2['Val'] = df1['ER_Shallow']
    df = pd.concat([df1, df2])
    depths = df['max_depth'].unique()   
    fig, axs = plt.subplots(len(depths), figsize=(15,10), )
    #fig.suptitle('Expansion Rate vs. Epsilon', fontsize=20)
    for j in range(len(depths)) :
        depth = depths[j]
        ax = sns.lineplot(ax = axs[j], x = 'epsilon', y = 'Val', data = df[df['max_depth']==depth], hue = 'Algorithm', ci = None, palette=pairs[j*2:j*2+2], marker='o')
        ax.set_xlabel(' ',labelpad=20)
        ax.set_ylabel('Expansion Rate', rotation=90, fontsize=20, labelpad=20)
        ax.set_ylim(0, 1)
        ax.set_title(f'Depth = {depth}', fontsize=20)
        ax.set_xticks(list(set(df['epsilon'])))
        plt.setp(ax.get_legend().get_texts(), fontsize='15') 
        plt.setp(ax.get_legend().get_title(), fontsize='15')
        plt.setp(ax.get_xticklabels(), fontsize=14)         
        plt.setp(ax.get_yticklabels(), fontsize=14)    
    
        
    axs[-1].set_xlabel('Epsilon', rotation=0, fontsize=20, labelpad=10)
    fig.tight_layout() 
    plt.show()
print('Max sum = 64, Branching factor = 5, NPlayers = 3')
plot_scatter3(output,output['epsilon'].unique())
```

```python
def plot_scatter4(df, epsilons):
    sns.set_theme()
    fig, axs = plt.subplots(1, figsize=(15,8))
    ax = sns.lineplot(ax = axs, x = 'epsilon', y = 'ER_Bounded', data = df, hue = 'max_depth', ci = None, palette=pal[:len(set(df['max_depth']))], marker='o')
    ax.set_ylim(0, 1)
    ax.set_title('Expansion Rate vs. Epsilon', fontsize=20)
    ax.set_xlabel('Epsilon', rotation=0, fontsize=20, labelpad=10)
    ax.set_ylabel('Expansion Rate', rotation=90, fontsize=20, labelpad=20)
    ax.set_xticks(list(set(df['epsilon'])))
    ax.legend().set_title('Depth')
    plt.setp(ax.get_legend().get_texts(), fontsize='20') 
    plt.setp(ax.get_legend().get_title(), fontsize='20')
    plt.setp(ax.get_xticklabels(), fontsize=14)         
    plt.setp(ax.get_yticklabels(), fontsize=14)         
print('Max sum = 64, Branching factor = 5, NPlayers = 3')
plot_scatter4(output,output['epsilon'].unique())
```

```python
def plot_scatter5(df, epsilons):
    depths = df['max_depth'].unique()   
    fig, axs = plt.subplots(len(depths), figsize=(15,10))
    fig.suptitle('Result Suboptimality vs. Epsilon', fontsize=20)
    x = df['epsilon']
    y_bound = df['Bound_Result_Gap']
    
    for j in range(len(depths)) :
        depth = depths[j]
        shallow_er_means = [4,8,16,24,32]
        bounded_er_means = []
        for i in df['epsilon'].unique():
            query = 'max_depth=={} '.format(depth)
            temp1 = df.query(query)
            query = 'epsilon=={} '.format(i)
            temp1 = temp1.query(query)
            bounded_er_means.append(np.mean(temp1['Bound_Result_Gap']))
        axs[j].plot(df['epsilon'].unique(), shallow_er_means, label='Max possible suboptimality', linestyle='-.')
        axs[j].plot(df['epsilon'].unique(), bounded_er_means, label='Actual result suboptimality. depth = {}'.format(depth), linestyle='--')
        axs[j].set_xlabel('epsilon', rotation=0, fontsize=15, labelpad=10)
        axs[j].set_ylabel('result suboptimality', rotation=90, fontsize=15, labelpad=20)
        axs[j].set_ylim(0, 35)
        axs[j].legend(loc='upper right', bbox_to_anchor=(1.42, 1.015),prop={'size': 14})
    
    plt.show()
print('Max sum = 64, Branching factor = 5, NPlayers = 3')
plot_scatter5(output,output['epsilon'].unique())
```

```python
def plot_scatter6(df, epsilons):
    
    sns.set_theme()
    fig, axs = plt.subplots(1, figsize=(15,10))
    ax = sns.lineplot(ax = axs, x = 'epsilon', y = 'Bound_Result_Gap', data = df, hue = 'max_depth', ci = None, palette=pal[:len(set(df['max_depth']))], marker='o')
    ax.set_ylim(-1, 35)
    ax.set_title('Final Score Difference vs. Epsilon', fontsize=20)
    ax.set_xlabel('Epsilon', rotation=0, fontsize=20, labelpad=10)
    ax.set_ylabel('Final Score Difference', rotation=90, fontsize=20, labelpad=20)
    ax.set_xticks(list(set(df['epsilon'])))
    ax.legend().set_title('Depth')
    ax.plot(df['epsilon'].unique(), df['epsilon'].unique(), label='Max possible suboptimality', linestyle='-.', color ='r', marker='o')
    ax.text(13, 13.7, 'Score Difference Bound', fontsize=25, rotation = 29)
    plt.setp(ax.get_legend().get_texts(), fontsize='20') 
    plt.setp(ax.get_legend().get_title(), fontsize='20')
    plt.setp(ax.get_xticklabels(), fontsize=14)         
    plt.setp(ax.get_yticklabels(), fontsize=14)         
print('Max sum = 64, Branching factor = 5, NPlayers = 3')
plot_scatter6(output,output['epsilon'].unique())
```

```python
def plot_scatter_bars(df, epsilons):
    x = df['epsilon']
    y_bound = df['ER_Bounded']
#     axs.scatter(x, y_bound, marker='x')    
    depths = df['max_depth'].unique()   
    eps =df['epsilon'].unique()
    for j in range(len(depths)) :
#         fig, axs = plt.subplots(1, figsize=(15,10))
        depth = depths[j]
        shallow_wins_rates = []
        bounded_wins_rates = []
        for i in eps:
            query = 'max_depth=={} '.format(depth)
            temp1 = df.query(query)
            query = 'epsilon=={} '.format(i)
            temp1 = temp1.query(query)
            
            shallow_wins_rates.append(np.count_nonzero(temp1['Root_Won_Shallow']))
            bounded_wins_rates.append(np.count_nonzero(temp1['Root_Won_Bounded']))
        index = [4, 8, 16, 24, 32]
        df_wins = pd.DataFrame({'bounded won': bounded_wins_rates,'shallow won': shallow_wins_rates}, index=index)
        axs = df_wins.plot.bar(rot=0)
        axs.set_title('Games Won vs. Epsilon : depth = {}'.format(depth), fontsize=15)
#         axs[j].plot(df['epsilon'].unique(), shallow_er_means, label='Mean shallow er depth = {}'.format(depth), linestyle='-.')
#         axs[j].plot(df['epsilon'].unique(), bounded_er_means, label='Mean bounded er depth = {}'.format(depth), linestyle='--')
        axs.set_xlabel('epsilon', rotation=0, fontsize=10, labelpad=10)
        axs.set_ylabel('games_won', rotation=90, fontsize=10, labelpad=20)
        axs.set_ylim(0, 50)
        axs.legend(loc='upper right', bbox_to_anchor=(1.43, 1.015))
    
        plt.show()
plot_scatter_bars(output,output['epsilon'].unique())
```
