```python
def generate_all3_vector_values( max_sum) : 
    all_vectors=[]
    for i in range(max_sum+1):
        for j in range(max_sum+1):
            for k in range(max_sum+1):
                if i+j+k==max_sum:
                    vector=[i,j,k]
                    all_vectors.append(vector)
    return all_vectors

def calc_leaf_value(node, all_vectors):
    global c1, c2
    res = 0
    curr = node
    while not curr.id=='Root':
        res += pow(curr.index + c1, curr.depth+c2)  #can change the calculate 
        curr = curr.parent
    node.value = all_vectors[res%len(all_vectors)]
#     print("index: {} , depth: {} , value: {} ".format(node.index,node.depth,node.value))
    return node

def generate_node(parent_node, child_index, max_depth, all_vectors):
    global n_players, nodes_expanded
    nodes_expanded+=1
    curr = AnyNode(id= '', parent=parent_node, index=child_index, depth=parent_node.depth+1, turn=(parent_node.depth+1)%n_players, value=np.array([0]*n_players))
    if curr.depth == max_depth:
        curr = calc_leaf_value(curr, all_vectors)
    return curr

def calc_tree_size(max_depth, branching_factor):
    res = 0
    for i in range(max_depth+1):
        res+= pow(branching_factor, i)
    return res
```

```python
def Bounded_Pruning2(node, bound, epsilon):
    global max_sum, promised, n_players, max_depth, all_vectors
    if not np.array_equal(node.value,np.array([0]*n_players)):
#         print('leaf found - value:', node.value, 'index:', node.index, 'depth:', node.depth, 'parent idx:', node.parent.index)
        return node.value, node.index
    b_best, index = Bounded_Pruning2(generate_node(node, 0, max_depth, all_vectors), max_sum, epsilon)
#     if node.id == 'Root':
#         print('returned: ', b_best, index)
#         return b_best,node.index
    #print("index : {}".format(index))
    for child in range(1, branching_factor):
        if node.id == 'Root' or node.id=='root':
            promised = max(promised, b_best[0])
#             print('promised try update :', promised)
#             print('b_best :', b_best)
            if promised + epsilon >= max_sum:
                if node.id in ['root','Root']:
                    return b_best, index
                else:
                    return b_best, node.index
#                 return best, node.index
        if node.turn == 1 and node.parent.id in ['root','Root'] and promised + epsilon >= max_sum - b_best[node.turn] and not promised==0:
#             print("prune - promised:",promised , 'epsilon:',epsilon , 'max_sum:', max_sum , 'b_best', b_best[node.turn])
            if node.id in ['root','Root']:
                return np.array([-1]+[0]*(n_players-1)), index
            else:
                return np.array([-1]+[0]*(n_players-1)), node.index
#             return np.array([-1]+[0]*(n_players-1)), node.index
        if (b_best[node.turn] >= bound):
            if node.id in ['root','Root']:
                return b_best, index
            else:
                return b_best, node.index
#             return b_best, node.index
        current , curr_index = Bounded_Pruning2(generate_node(node, child, max_depth, all_vectors), max_sum-b_best[node.turn], epsilon)
        #print("curr_index : {}".format(curr_index))
        if current[node.turn] > b_best[node.turn]:
            #print("enter")
            b_best = current
            index = child
    if node.id in ['root','Root']:
        return b_best, index
    else:
        return b_best, node.index
#     return b_best, node.index
    
def Bounded_Pruning_Root2(node, bound, epsilon):
    global promised
    promised=0
#     if not node.id=='Root':
#         node.id = 'root'
    return Bounded_Pruning2(node, bound, epsilon)

def Shallow2(node, bound):
    global max_sum, n_players, max_depth, all_vectors
    if not np.array_equal(node.value,np.array([0]*n_players)):
        return node.value, node.index
    best, index = Shallow2(generate_node(node, 0, max_depth, all_vectors), max_sum)
    for child in range(1, branching_factor):
        if best[node.turn]>=bound:
            if node.id in ['root','Root']:
                return best, index
            else:
                return best, node.index
#             return best, node.index
        current, curr_index = Shallow2( generate_node(node, child, max_depth, all_vectors), max_sum- best[node.turn])
        if current[node.turn] > best[node.turn]:
            best = current
            index = child
    if node.id in ['root','Root']:
        return best, index
    else:
        return best, node.index
#     return best, node.index
```

```python
def simulate_with_bounded():
    global n_players, max_depth, max_sum, epsilon, all_vectors
    root = AnyNode(id='Root', index=0, depth=0, turn=0, value=np.array([0]*n_players))
    game_over = False
    player_index = 0
    curr= root
    while not game_over:
        if curr.depth == max_depth:
            game_over = True
            final_score = curr.value
            break
        if player_index==0:
            curr_score, chosen_index = Bounded_Pruning_Root2(curr, max_sum, epsilon)
        else:
            curr_score, chosen_index = Shallow2(curr, max_sum)
        print('Player ', player_index, 'selected ', chosen_index, 'expected score:', curr_score)
        if not curr.id=='Root':
             curr.id=''
        curr = generate_node(curr, chosen_index, max_depth, all_vectors)
        curr.id = 'root'
        player_index = (player_index+1)%n_players
    print("with bounded: " + str(final_score))
    return final_score


def simulate_only_shallow():
    global n_players, max_depth, max_sum, epsilon, all_vectors
    root = AnyNode(id='Root', index=0, depth=0, turn=0, value=np.array([0]*n_players))
    game_over = False
    player_index = 0
    curr= root
    while not game_over:
        if curr.depth == max_depth:
            game_over = True
            final_score = curr.value
            break
        curr_score, chosen_index = Shallow2(curr, max_sum)
#         print('Player ', player_index, 'selected ', chosen_index, 'expected score:', curr_score)
        if not curr.id=='Root':
             curr.id=''
        curr = generate_node(curr, chosen_index, max_depth, all_vectors)
        curr.id = 'root'
        player_index = (player_index+1)%n_players
    print("only shallow: " + str(final_score))
    return final_score
```

```python
epsilons = [4,8,16,24,32]
max_sum = 64
max_depths = [4,7,10]
n_players = 3
branching_factor = 5
returns = 50
for max_depth in max_depths:
    nplayers_res, maxsum_res, epsilon_res, max_depth_res, branching_res, bound_simu_res, bound_res, opt_res, bound_expended, opt_expended, tree_size, bound_wins, opt_wins, sour_losers, bound_result_suboptimality =  [], [], [], [], [], [], [], [], [], [], [], [], [], [], []
    for epsilon in epsilons:
        c1, c2 = 2, 3
        for i in range(returns):                                                           
            promised = 0
            nplayers_res.append(n_players)
            maxsum_res.append(max_sum)
            epsilon_res.append(epsilon)
            max_depth_res.append(max_depth)
            branching_res.append(branching_factor)
            all_vectors =  generate_all3_vector_values(max_sum)
            root = AnyNode(id='Root', index=0, depth=0, turn=0, value=np.array([0]*n_players))
            print('max depth:{}, epsilon:{}, index:{}/{}'.format(max_depth,epsilon, i, returns))
    
            tree_size = calc_tree_size(max_depth, branching_factor)
            nodes_expanded = 1
            b = Bounded_Pruning_Root2(root,max_sum,epsilon)[0]
            bound_expended.append(nodes_expanded/tree_size)
            nodes_expanded = 1
            s = Shallow2(root,max_sum)[0]
            opt_expended.append(nodes_expanded/tree_size)
    
            print('game simu:')
            bound_simu_res.append( simulate_with_bounded())
            bound_res.append(b)
            opt_res.append(s)    
            bound_win = b.index(max(b)) == 0
            opt_win = s.index(max(s)) == 0
            bound_wins.append(bound_win)
            opt_wins.append(opt_win)
            sour_losers.append(opt_win and not bound_win)
            bound_result_suboptimality.append(s[0]-b[0])
            c1 +=1
            c2 +=1
    df = pd.DataFrame()
    df['n_players']= pd.Series(nplayers_res)
    df['max_sum']= pd.Series(maxsum_res)
    df['epsilon']= pd.Series(epsilon_res)
    df['max_depth']= pd.Series(max_depth_res)
    df['branching_factor']= pd.Series(branching_res)

    df['Shallow_Score']= pd.Series(opt_res)
    df['Bounded_Score']= pd.Series(bound_res)
    df['Bounded_Simulation_Score']= pd.Series(bound_simu_res)
    df['ER_Bounded']= pd.Series(bound_expended)
    df['ER_Shallow']= pd.Series(opt_expended)
    df['Root_Won_Bounded']= pd.Series(bound_wins)
    df['Root_Won_Shallow']= pd.Series(opt_wins)
    df['Sour_Loser']= pd.Series(sour_losers)
    df['Bound_Result_Gap']= pd.Series(bound_result_suboptimality)
    df.to_csv(r'C:\Users\onagar\OneDrive - Intel Corporation\Desktop\bounded_test_output_{}.csv'.format(max_depth))
```
